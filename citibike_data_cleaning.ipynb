```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Importing data files

bike_2019_df = pd.read_csv("resources/201903-citibike-tripdata.csv")
bike_2020_df = pd.read_csv("resources/202003-citibike-tripdata.csv")
bike_2018_df = pd.read_csv("resources/201803-citibike-tripdata.csv")

bike_2019_df.head()
```

```python
bike_2018_df.info()
```

```python
# Changing the column to correct datetime format
bike_2019_df['starttime'] = pd.to_datetime(bike_2019_df['starttime'])
bike_2020_df['starttime'] = pd.to_datetime(bike_2020_df['starttime'])
bike_2018_df['starttime'] = pd.to_datetime(bike_2018_df['starttime'])
bike_2019_df['starttime']
```

```python
# Changing the column to correct datetime format
bike_2018_df['starttime'] = pd.to_datetime(bike_2018_df['starttime'])
bike_2018_df['starttime']
```

```python
# Changing the column to correct datetime format
bike_2020_df['stoptime'] = pd.to_datetime(bike_2020_df['stoptime'])
bike_2020_df['stoptime']
```

```python
bike_2018_df.info()
```

```python
bike_2019_df = bike_2019_df.dropna()
bike_2020_df = bike_2020_df.dropna()
bike_2018_df = bike_2018_df.dropna()
```

```python
bike_2019_df = bike_2019_df.fillna(0)
bike_2020_df = bike_2020_df.fillna(0)
bike_2018_df = bike_2018_df.fillna(0)
```

```python
# Converting decimal values to integer
bike_2019_df['start station id'] = bike_2019_df['start station id'].astype(np.int64)
bike_2019_df['end station id'] = bike_2019_df['end station id'].astype(np.int64)
bike_2020_df['start station id'] = bike_2020_df['start station id'].astype(np.int64)
bike_2020_df['end station id'] = bike_2020_df['end station id'].astype(np.int64)
bike_2018_df['start station id'] = bike_2018_df['start station id'].astype(np.int64)
bike_2018_df['end station id'] = bike_2018_df['end station id'].astype(np.int64)
```

```python
# Renaming columns
column_names = list(bike_2019_df.columns)
column_names[0:14] = ['trip_duration', 'start_time', 'stop_time', 'start_station_id', 'start_station_name', 
                      'start_station_latitude', 'start_station_longitude',  
                      'end_station_id', 'end_station_name', 'end_station_latitude', 
                      'end_station_longitude', 'bike_id', 'user_type', 'birth_year']
bike_2019_df.columns = column_names

bike_2019_df.head()
```
